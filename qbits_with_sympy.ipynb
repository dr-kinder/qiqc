# Quantum Math with Sympy

This notebook introduces some tools from the Python library SymPy for quantum computing.  SymPy is a library for symbolic math.  These quantum tools are a small subset, but useful for our course.

```python
# Import some quantum computing tools.
# First, basic math and symbols
from sympy import symbols, cos, sin, pi, I, exp, sqrt
from sympy import simplify, trigsimp, expand, evalf, Q

# Next, quantum algebra
from sympy.physics.quantum import represent, qapply, Bra, Ket, Dagger

# Now, quantum computing
from sympy.physics.quantum.qubit import Qubit
from sympy.physics.quantum.gate import X,Y,Z,H,S,T
from sympy.physics.quantum.gate import IdentityGate as Id

# Print nice expressions as output.
from sympy import init_printing
init_printing()
```

## Qbits

Creating qbits and gates is simple.  Since SymPy is a symbolic libarary,  you have to be explicit about asking to see results.  `qapply()` takes care of this.

```python
# Define a qubit in the |0> state
q = Qubit('0')
display(q)
```

We can assign variables to specific qbits.

```python
psi = (Qubit(0) + I*Qubit(1)) / sqrt(2)
display(psi)
```

We can do standard quantum operations on them.

```python
Dagger(psi)
```

```python
Dagger(psi)*psi
```

```python
qapply(Dagger(psi)*psi)
```

We can even do projection.

```python
qapply(Dagger(psi)*Qubit(0))
```

## Gates

We can also do algebra with gates.  When defining a single qbit gate, we have to specify which qbit in a register to act on.  If there is only one qbit, this should be '0'.

```python
h = H(0)
represent(h,nqubits=1)
```

We can apply gates to qbits, of course.

```python
# Apply a Hadamard gate (H) to put it in superposition
state = H(0) * Qubit(0)
display(state)

# Evaluate the expression
qapply(state)
```

We can build gates by combining the standard set.

```python
U = H(0) * X(0) * H(0)
represent(U,nqubits=1)
```

```python
represent(X(0)*Y(0), nqubits=1)
```

## Symbolic Algebra

SymPy is made to work with symbols, so we can do some pretty fun stuff.

```python
A, x, theta, phi = symbols('A x theta phi', real=True)
alpha, beta = symbols('alpha beta', complex=True)

q = A * (Qubit(0) + x * Qubit(1))
psi = alpha * Qubit(0) + beta * Qubit(1)

display(q)
display(psi)
```

```python
display(simplify(qapply(Dagger(q) * q)))
display(simplify(qapply(Dagger(psi) * psi)))
```

```python
a = cos(theta/2)
b = sin(theta/2) * exp(I*phi)
display(a)
display(b)
```

```python
b*b.conjugate()
```

```python
z = a + b
print("Start with a complex symbolic expression.")
display(z)
print("SymPy knows Euler's identity.")
display(expand(z,complex=True))
print("Compute the modulus squared |z|┬▓.")
display(z*z.conjugate())
print("Expand ...")
display(expand(z*z.conjugate(),complex=True))
print("... and simplify.  (SymPy knows trig identities, too.)")
display(trigsimp(expand(z*z.conjugate(), complex=True)))
```
